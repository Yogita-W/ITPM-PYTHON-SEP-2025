```python
#Wap to check given number is even or odd.
def Even_odd(num):
    if num%2==0:
        return f"{num} is even"
    
    else : 
        return f"{num} is Odd"
    
Even_odd(23)    
```

```python
#addition of two numbers.
add = lambda x,y : x+y
print(add(10,20))
```

```python

max = lambda x : x
print(max([10,20,30]))
```

```python
string = lambda x : x if x == x[::-1] else "False"
print(string("ab"))
```

```python
String1 = lambda x : "Palimdrom" if x == x[::-1] else  "Not"
print(String1("hello"))
print(String1("madam"))
```

```python
String1 = lambda x : f"{x} Palimdrom String" if x == x[::-1]  else  f"{x} Not "
print(String1("hello"))
print(String1("madam"))
```
