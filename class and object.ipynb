```python
class Calculator:
    def add(self,a,b):
        self.a = a
        self.b = b
        return f"Addition of {a} and {b} : {a+b}"
    def Sub(self,a,b):
        self.a = a
        self.b = b
        return f"Addition of {a} and {b} : {a-b}"
    def Mul(self,a,b):
        self.a = a
        self.b = b
        return f"Addition of {a} and {b} : {a*b}"
    def Div(self,a,b):
        self.a = a
        self.b = b
        return f"Addition of {a} and {b} : {a/b}"
    
obj1 =  Calculator()
print(obj1.add(20,10))
print(obj1.Sub(20,10))
print(obj1.Mul(20,10))  
print(obj1.Div(20,10))
    
    
    
```

```python
class calculator:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        return f"Addition of {self.a} and {self.b} is : {self.a+self.b}"
    def Sub(self):
        return f"Addition of {self.a} and {self.b} is : {self.a-self.b}" 
    def Mul(self):
        return f"Addition of {self.a} and {self.b} is : {self.a*self.b}"
    def Div(self):
        return f"Addition of {self.a} and {self.b} is : {self.a/self.b}"
obj1 = calculator(20,10)
print(obj1.add())
print(obj1.Sub())
print(obj1.Mul())
print(obj1.Div())        
```

```python
class Car:
    def Name(self,name):
        self.name = name
        return f"{name} This is my car"
    def Color(self,color):
        self.color = color
        return f"Car's Color is :{color}"
    def Engin(self,x):
        self.x = x
        return f"Engin is : {x} CC"
obj1 = Car()
print(obj1.Name("BMW"))
print(obj1.Color("Black"))
print(obj1.Engin("350"))    
```

```python
class Animal :
    def Name (self,name):
        self.name=name
        return f"This is {name}"
    def Eat(self,food):
        self.food = food
        return f"{food} this is his/her fav.food"
    
obj1 = Animal()
print(obj1.Name("Tiger"))
print(obj1.Eat("Deer"))    

obj2 = Animal()
print(obj2.Name("Lion"))
print(obj2.Eat("Deer"))
```
