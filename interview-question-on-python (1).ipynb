```python
#2.Calculate the Sum of Digits of a Given Number
num=int(input("Enter the number:"))
sum_of_digits = 0
while num > 0:
  last_digit = num % 10
  sum_of_digits += last_digit
  num //= 10
print(sum_of_digits)
```

```python
#5.Find the Greatest Common Divisor (GCD) of Two Numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
gcd = 1
i = 1 
while i <= min(num1, num2):  
  if num1 % i == 0 and num2 % i == 0:  
    gcd = i  
  i += 1
print(f"The GCD of {num1} and {num2} is: {gcd}")
```

```python
#7. Check if a Number is a Palindrom
num=int(input("Enter the number:"))
reverse_num=0
original_num= num
while num>0 :
    reverse_num = reverse_num * 10 + num % 10
    num //= 10
if reverse_num == original_num:
  print("The number is Palindrome")
else:
  print("The number is not Palindrome")
```

```python
#10.	Bank Account Example:
#o	Create a BankAccount base class with methods like deposit() and withdraw().
#o	Create a SavingsAccount class that inherits from BankAccount and adds an attribute interest_rate.
#o	Create a CheckingAccount class that inherits from BankAccount and adds an attribute overdraft_limit.


class BankAccount:
    def __init__(self, account_number, balance=0.0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Insufficient funds or invalid amount.")

    def display_balance(self):
        print(f"Account {self.account_number} balance: {self.balance}")

class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance=0.0, interest_rate=0.02):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def apply_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest applied: {interest}. New balance: {self.balance}")

class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance=0.0, overdraft_limit=500):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if 0 < amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Overdraft limit exceeded or invalid amount.")
            
savings = SavingsAccount("SA123", 1000, 0.05)
savings.display_balance()
savings.apply_interest()
savings.withdraw(200)

checking = CheckingAccount("CA456", 500, 1000)
checking.display_balance()
checking.withdraw(1200)
checking.withdraw(5000) 
```

```python
#12.	Shape Example:
#o	Create a Shape base class with a method calculate_perimeter().
#o	Create a Rectangle class that inherits Shape and implements the perimeter calculation.
#o	Create a Triangle class that inherits Shape and implements the perimeter calculation.

class Shape:
    def calculate_perimeter(self):
        raise NotImplementedError("Subclasses must implement calculate_perimeter method")

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

class Triangle(Shape):
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
    
    def calculate_perimeter(self):
        return self.side1 + self.side2 + self.side3

rectangle = Rectangle(10, 5)
print("Rectangle perimeter:", rectangle.calculate_perimeter())

triangle = Triangle(3, 4, 5)
print("Triangle perimeter:", triangle.calculate_perimeter())
```

## Pattern Programming

```python
# Right angle triangle
rows = 5
for i in range(1, rows + 1):
    for j in range(i):
        print('*', end='')
    print()
```

```python
#Inverted Right-Angled Triangle Pattern#
rows = 5
for i in range(rows, 0, -1):  
    for j in range(i):  
        print('*', end='')  
    print()  
```

```python
#3.Number Triangle:
rows = 5  # Number of rows

for i in range(1, rows + 1):  
    for j in range(1, i + 1):  
        print(j, end='')  
    print() 
```

```python
#4.Same Number Row
rows = 5
for i in range(1, rows + 1):
    for j in range(i):
        print(i, end='')
    print() 

```

```python
#5.Reverse Number Triangle
rows = 5  

for i in range(rows, 0, -1):  
    for j in range(rows, rows - i, -1):  
        print(j, end='')  
    print()  
```

```python
#6. pyramid
rows = 5  

for i in range(1, rows + 1):  
    for j in range(rows - i):  
        print(" ", end="")   
    for k in range(2 * i - 1):  
        print("*", end="") 
    print()   
```

```python
#7.Inverted pyramid
rows = 5  

for i in range(rows, 0, -1):  
    for j in range(rows - i):  
        print(" ", end="")   
    for k in range(2 * i - 1):  
        print("*", end="")    
    print()    
```

```python
#8.Diamond
rows = 5  

# Upper part of the diamond
for i in range(1, rows + 1):  
    for j in range(rows - i):  
        print(" ", end="")  
    for k in range(2 * i - 1):  
        print("*", end="")  
    print()  

# Lower part of the diamond
for i in range(rows - 1, 0, -1):  
    for j in range(rows - i):  
        print(" ", end="")  
    for k in range(2 * i - 1):  
        print("*", end="")  
    print()
```

```python
#9.Number Pyramid
rows = 5  

for i in range(1, rows + 1):  
    # Print spaces
    for j in range(rows - i):  
        print(" ", end="")  

    # Print increasing numbers
    for k in range(1, i + 1):  
        print(k, end="")  

    # Print decreasing numbers
    for l in range(i - 1, 0, -1):  
        print(l, end="")  

    print()  
```

```python
#10. Hollow Square
rows = 5  
cols = 5  

for i in range(rows):  
    for j in range(cols):  
        if i == 0 or i == rows - 1 or j == 0 or j == cols - 1:  
            print("*", end="")   
        else:  
            print(" ", end="")   
    print()    
```

```python
#11.Floyd's Triangle
rows = 5  
num = 1  # Starting number  

for i in range(1, rows + 1):  
    for j in range(i):  
        print(num, end=" ")  
        num += 1    
    print()   
```

```python
#12.Pascal's Triangle
from math import factorial  

rows = 5  

for i in range(rows):  
    for j in range(rows - i - 1):  
        print(" ", end="")    
    for j in range(i + 1):  
        print(factorial(i) // (factorial(j) * factorial(i - j)), end="")
    print()
```

```python
#13. Butterfly 
rows = 5  

# Upper half
for i in range(1, rows + 1):  
    for j in range(i):  
        print("*", end="")  
    for j in range(2 * (rows - i)):  
        print(" ", end="")   
    for j in range(i):  
        print("*", end="")  
    print()  

# Lower half
for i in range(rows - 1, 0, -1):  
    for j in range(i):  
        print("*", end="")    
    for j in range(2 * (rows - i)):  
        print(" ", end="")
    for j in range(i):  
        print("*", end="")    
    print()
```

```python
# 14. Zig-zag pattern
rows = 3  
cols = 9  

for i in range(rows):  
    for j in range(cols):  
        if (i + j) % 4 == 0 or (i == 1 and j % 4 == 2):  
            print("*", end="")  
        else:  
            print(" ", end="")  
    print()  
```

```python
#15.Checkerboard Pattern
rows = 8  
cols = 8  

for i in range(rows):  
    for j in range(cols):  
        if (i + j) % 2 == 0:  
            print("*", end=" ")    
        else:  
            print(" ", end=" ")    
    print()   
```

```python
#18.Hourglass
rows = 5  


for i in range(rows, 0, -1):  
    print(" " * (rows - i), end="")  
    print("*" * (2 * i - 1), end="")  
    print()  


for i in range(2, rows + 1):  
    print(" " * (rows - i), end="") 
    print("*" * (2 * i - 1), end="") 
    print()  
```

```python
#19.Wave pattern
rows = 4  
cols = 10  

for i in range(rows):  
    for j in range(cols):  
        if (j % 4 == 0 and i == 2) or (j % 4 == 1 and i == 1) or (j % 4 == 2 and i == 0) or (j % 4 == 3 and i == 1):  
            print("*", end="")  
        else:  
            print(" ", end="")  
    print()  
```
