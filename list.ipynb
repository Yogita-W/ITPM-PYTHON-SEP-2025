```python
#Wap to print all Vowels in give string.
str1 = "Python Programming"
vowels = "aeiouAEIOU"
v_count = 0
vowel_list = []
for i in str1:
    if i in vowels:
        v_count+=1
        vowel_list.append(i)
print(v_count)
print(vowel_list)        
```

```python
str1 = "Python Programming"
vowels = "aeiouAEIOU"
v_count = 0
vowel_list = []
c_count = 0
c_list = []
for i in str1:
    if i in vowels:
        v_count+=1
        vowel_list.append(i)
    else :
        c_count+=1
        c_list.append(i)    
print(v_count)
print(vowel_list) 
print(c_count)
print(c_list)    
```
