```python
class Calculator:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
        
    def Add(self):
        return f"Addition of {self.a,self.b,self.c} :{self.a+self.b+self.c}"
    
obj1 = Calculator(12,32,44)
obj1.Add()        
```
