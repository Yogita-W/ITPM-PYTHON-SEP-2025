```python
class Num:
    def Numbers(self,a,b):
        self.a = a
        self.b = b
        return f"a:{a} and b:{b}"
class Add:
    def Addition(self,a,b):
        self.a = a
        self.b = b
        return f"Addition of a and b : {a+b}"
            
    
    
obj = Num()
print(obj.Numbers(10,20))
obj2 = Add() 
print(obj2.Addition(10,20))  
```

```python
class Number:
    def num(self,a,b):
        self.a = a
        self.b = b
        return f"{a} and {b}"
class Add(Number):
    def add(self):
        return f"Addition of a and b is : {self.a+self.b}"   
        

obj1 = Add()
obj1.num(10,20)
obj1.add()
```

```python
class Parent:
    def home(self,name):
        self.name = name
        return f"{name} This is My home"
    
class child(Parent):
    def home2(self):
        return f"{self.name} this is my Father's home"
    
obj1 = child()
print(obj1.home("Matoshri"))
print(obj1.home2() )
        
```

```python
class Contact:
    def Number(self,n1,n2):
        self.n1 = n1
        self.n2 = n2
        
class whatsup(Contact):
    def Use(self):
        return f"{self.n1} {self.n2}"  
    
obj1 = whatsup()
obj1.Number(9876543223,9087654323) 
print(obj1.Use())       
```

```python
class Parent:
    def Mobile(self,use):
        self.use = use
    
class child(Parent):
    def Mobile2(self,use1):
        self.use1 = use1
        return f"{self.use} This is my mother's Phone and {use1} this is my phone "    
        
obj1 = child()
obj1.Mobile("Redme")
print(obj1.Mobile2("iphone"))        
```

```python
#Multiple Inheritance
class Parent1:
    def name1(self,n1):
        self.n1 = n1
        
class Parent2:
    def name2(self,n2):
        self.n2 = n2
        
class Child(Parent1,Parent2):
    def name(self,n):
        self.n = n
        return f"My name is {n} {self.n2} and my mothers name is {self.n1}"   
    
obj1 = Child()
obj1.name1("Lalita Mahabole")
obj1.name2("Santosh Mahabole")
obj1.name("om")               
```
