## Encapsulation :
Encapsulation is one of the fundamental concepts in object-oriented programming (OOP). **It describes the idea of wrapping data and the methods that work on data within one unit.** This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data.
A class is an example of encapsulation as it encapsulates all the data that is member functions, variables, etc. **The goal of information hiding is to ensure that an object’s state is always valid by controlling access to attributes that are hidden from the outside world**.
Encapsulation also restricts direct access to some components of an object, which is known as data hiding.

## Key Features of Encapsulation:
### Access Control:

**Use access modifiers to control visibility.**

**Public:** Accessible from anywhere (attribute_name).

**Protected:** Shouldn't be accessed outside the class or subclass (_attribute_name).

**Private:** Only accessible within the class (__attribute_name).

**Getter and Setter Methods:**

Encapsulation allows defining methods to access and modify private attributes safely.


### Public Access Modifier:
The members of a class that are declared public are **easily accessible from any part of the program.** All data members and member functions of a class are public by default.

Syntax: No leading underscores (e.g., variable or method()).

Access: Public members are accessible from anywhere: inside the class, outside the class, and even in derived classes.

Default behavior: All class members are public by default in Python.

### Protected members
****Protected members (in C++ and JAVA) are those members of the class that cannot be accessed outside the class but can be accessed from within the class and its subclasses.** ** To accomplish this in Python, just follow the convention by prefixing the name of the member by a** single underscore “_”**.

Although the protected variable can **be accessed out of the class as well as in the derived class (modified too in derived cla**ss), it is customary(convention not a rule) to not access the protected out the class bod

Syntax: One leading underscore (e.g., _variable or _method()).

Access: Protected members are intended to be accessible only within the class and its subclasses. However, this is just a convention; they are still accessible from outside the class but indicate that you should not do so unless necessary.

Use case: These members are meant for internal use or use in derived classes.y.



```python
SELF.NAME = NAME #PUBLIC(ANY ONE CAN ACCESS)
SELF._NAME = NAME #PROTECTED(ONLY CHILD CLASS acess)
SELF.__NAME = NAME # PRIVETE(no one )
```

```python
class Example:
    def __init__(self):
        self._protected_var = "I am protected" #protected 

    def _protected_method(self):
        return "This is a protected method"

class SubExample(Example):
    def access_protected(self):
        return self._protected_var

obj = Example()
print(obj._protected_var)  # Accessible but discouraged
print(obj._protected_method())  # Accessible but discouraged

sub_obj = SubExample()
print(sub_obj.access_protected())  # Proper usage
```

### Private members****Private members are similar to protected members, but the difference is that class members declared as private cannot be accessed outside the class, nor by any derived class.****. In other words, private members are only accessible within the class where they are declared.

In Python, there is no strict enforcement of private instance variables that cannot be accessed from outside the class. However, by convention, private variables are indicated by prefixing the variable name with an underscore (e.g., _variable or __variable). This indicates to developers that these variables are intended to be private, even though they can still be accessed if needed.

Howe**to define a private member prefix the member name with **double underscore “__”.****

Private
Syntax: Two leading underscores (e.g., __variable or __method()).

Access: Private members are not accessible directly from outside the class. Python uses name mangling to make these members less accessible by changing their names internally (e.g., __variable becomes _ClassName__variable).

Use case: These members are meant strictly for internal use within the class.



“__”.

```python
class Itpm:
    def __init__(self):
        self.__course = "DA"
        self.__tech = "Python"

    # Getter methods
    def get_course(self):
        return self.__course

    def get_tech(self):
        return self.__tech

    # Setter methods
    def set_course(self, course):
        self.__course = course

    def set_tech(self, tech):
        self.__tech = tech


# Object creation
obj = Itpm()

# Using getters
print(obj.get_course())
print(obj.get_tech())

# Using setters
obj.set_course("Data Analytics")
obj.set_tech("Python + Power BI")

# After update
print(obj.get_course())
print(obj.get_tech())
```
