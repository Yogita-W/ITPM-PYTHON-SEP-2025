```python
class Book:
    def __init__(self,title,author,price):
        self.title = title
        self.author = author
        self.price = price
    def Display(self):
        return f"Book Title :{self.title} Author : {self.author} Orignal Price:{self.price}"
    def Discount(self):
        discount = self.price * 0.10
        new_price = self.price -discount
        return f"10% Discount Price is : {discount} and New Price : {new_price}"
        
        
        
obj1 = Book("Python","Gaudo van rossum",2000)
print(obj1.Display())      
print(obj1.Discount())
```

```python
#2.Create a Shape class with a method display(). 
#Create a Circle class that inherits Shape and calculates the area of a circle.

class Shape:
    def __init__(self,r):
        self.r = r
    def display(self):
        return f" Radius of Circle : {self.r}"
class Circle(Shape):
    def area(self):
        return f" Area of Circle : {3.14*self.r*self.r}"
    
class Perimeter(Shape):
    def perimeter(self):
        return f"Perimeter of Circle is : {2*3.14*self.r}"    
    
    
obj1 = Circle(2)
print(obj1.display())
print(obj1.area())  
obj2 = Perimeter(2)

print(obj2.perimeter())  
              
```

Method Overriding : when we have same methods in parent class and child class at that time child class override functionality of parent class and display its own properties.

```python
class Emp:
    def details(self,name,salary):
        self.name = name
        self.salary = salary
        return f" am Emp and my name is {name} and my salary is {salary}"
class manager(Emp):
    def details(self,name,salary):
        self.name = name
        self.salary = salary
        return f" am Manager and my name is {name} and my salary is {salary}"   
    
obj1 = manager()
obj1.details("rohan",50000)     
```

Super () : to solve method overriding probleam we are using super ()

```python
class Emp:
    def details(self,name,salary):
        self.name = name
        self.salary = salary
        return f" am Emp and my name is {name} and my salary is {salary}"
class manager(Emp):
    def details(self,name,salary):
        emp_details = super().details("Om",4000)
        self.name = name
        self.salary = salary
        print(f" am Manager and my name is {name} and my salary is {salary}\n {emp_details}")  
    
obj1 = manager()
obj1.details("rohan",50000)     
```

Encapsulation: 

```python
from abc import ABC,abstractmethod
class Mobile(ABC):
    @abstractmethod
    def Green(self):
        pass
    @abstractmethod
    def Red(self):
        pass
class details(Mobile):
     def Green(self):
         return f"Recive call" 
     def Red(self):
         return f"Reject call"
     
obj1 = details()
print(obj1.Green())
print(obj1.Red())       
```

```python
from abc import ABC ,abstractmethod
class Login(ABC):
    
    @abstractmethod
    def email(self):
        pass
    
    @abstractmethod
    def password(self):
        pass
    
class user(Login):
    def email(self,e):
        self.e = e
        return f"email is correct"
    def password(self,password):
        self.password = password
        return f"password is correct"
    
obj1 = user()
print(obj1.email("rohan@gmail.com"))
print(obj1.password(1234))    
    
        
    
```
